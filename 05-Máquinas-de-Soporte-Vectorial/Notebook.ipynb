```python
# Importar dataset de la nube
from sklearn import datasets

datos = datasets.load_breast_cancer(as_frame = True)
datos.frame.head()
```

```python
# Información de la construcción de los datos
print(datos.frame.info())
```

```python
# Descripción estadística del contenido
datos.frame.describe()
```

```python
# Dividir datos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X = datos.data    # características
y = datos.target  # etiquetas

X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size = .25)  # 25% para la prueba
```

```python
from sklearn import svm, metrics

# Parámetros a iterar
kernels = ["linear", "rbf", "sigmoid"]
gammas = [1, .01, .001, .0001, .00001]

# Encontrando los parámetros óptimos
for kernel in kernels:
    for gamma in gammas:
        # Creación del modelo y segmentación de datos
        modelo = svm.SVC(kernel=kernel, gamma=gamma)
        modelo.fit(X_ent, y_ent)
        predicciones = modelo.predict(X_pru)

        # Mostrar resultados
        print(f"{kernel}: Gamma: {gamma}: Exactitud:", metrics.accuracy_score(y_pru, predicciones))
        print(f"{kernel}: Gamma: {gamma}: Precisión:", metrics.precision_score(y_pru, predicciones))
```

```python
# Modelo óptimo de máquinas de soporte vectorial
modelo = svm.SVC(kernel="linear")
modelo.fit(X_ent, y_ent)
predicciones = modelo.predict(X_pru)
```

```python
# Reporte de la clasificación
from sklearn.metrics import classification_report

print(classification_report(y_pru, predicciones))
```

```python
# Matriz de confusión
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pru, predicciones)
```

```python
# Características que más aportan al modelo
import numpy as np
coeficientes = modelo.coef_[0]
indices_importantes = np.argsort(np.abs(coeficientes))[::-1] # Ordenando por importancia

print("Características más importantes: \n")
for i in indices_importantes:
    print(f"{i} {X.columns[i]} =   \t{coeficientes[i]}")
```
